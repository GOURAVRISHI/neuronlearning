```python
df= pd.DataFrame(np.random.randn(100,3), columns= ['A', 'B', 'C'])
df.head()
df['A']= df['A'].cumsum()+20
df['B']= df['B'].cumsum()+20
df['C']= df['C'].cumsum()+20
```

```python
df.describe()
```

```python
# can download pic of any format and compare data on hover- amazing feature
# we can also import this data to plotly website and can make changes there 
# with changes we can also visualize same time.  
df.iplot()
```

```python
df.iplot(x='A', y='B', mode='markers', size=15)
```

```python
# important for outliers
# start with minimum 
# 25 %, 50%, 75% also refered as 1st quartile, 2nd quartile, 3rd quartile

# ex- 1, 2,3 8, 10, 15 ,20 - median - 8 , 
# 1st quartile= 2, 2nd quartile = 8, 3rd quartile= 15, 4th quartile= 20
# range of 1 to 3 rd quartile is interquartile range = 15-2=13
# to calculate min - 1st quartile-1.5 IQR (interquartile) = 2-1.5(13)= -17.5 beyond this is a outlier 
# to calculate max - 3st quartile+1.5 IQR (interquartile) = 15+1.5*13= 34.5 beyond this is a outlier 

df.iplot(kind='box')
```
