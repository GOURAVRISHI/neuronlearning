```python
# Merging by a common key
left = pd.DataFrame({ 'key':['K0', 'K8', 'K2', 'K3'],
                        'A':['A0', 'A1', 'A2', 'A3'],  
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({ 'key':['K0', 'K1', 'K2', 'K3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
print(left)
print("\n",right)
```

```python
# Inner is like Venn dia inner/ common intersection
merge1=pd.merge(left, right, how='inner', on='key')  
print("nAfter merging with 'inner' method\n", '-'*50)
print(merge1)
```

```python
# Outer is like Venn dia outside- area except common intersection
df1 = pd.DataFrame({ 'key1':['K0', 'K8', 'K2', 'K3'],
                        'A':['A0', 'A1', 'A2', 'A3'],  
                        'B': ['B0', 'B1', 'B2', 'B3']})

df2 = pd.DataFrame({ 'key2':['K0', 'K1', 'K2', 'K3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
print(left)
print("\n",right)
merge2=pd.merge(df1, df2, how='inner', left_on='key1', right_on='key2') 
print(merge2)
```
